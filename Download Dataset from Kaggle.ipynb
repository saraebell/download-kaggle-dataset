```python
# download Tesla Stock Price 
od.download("https://www.kaggle.com/datasets/rpaguirre/tesla-stock-price")
```

# 6. Unzip files (only need to do this step if data is donwloaded as zip files)

```python
# unzip files  
import zipfile

with zipfile.ZipFile("/content/tesla-stock-price.zip", 'r') as zip_ref:
    zip_ref.extractall("/content")
```
